<h1>Creating Balanced CICIoT 2023 Dataset</h1>
<h2>Background story</h2>
<p>We are working on a project where we are trying to use famous CICIoT 2023 dataset to detect the network attacks on IoT devices using Classical and moderan ML models keeping following constraint in mind:</p>
1. Minimum Computational power for Model.<br>
2. Minimum Size of model.

<h2>Objective</h2>
To get Balanced data for ML model to predict the intrusion attack more efficiently.

<h1>Balancing Strategy</h1>
We have to balance the data for all the target labels.<br>
There are different way of doing it.<br>
1) We balance according to the lowest label count.
2) We balance according to lowest Class count.

<h2>Label Count</h2>
<p>In label count we have Uploading_Attack with 1253 rows. So we have to reduce all other attacks to the same number of rows. </p>

<h2>ClassWise Count</h2>
In class count, we have Bruteforce class with lowest count. So we have to bring all other classes to this count.

```python
attacks=['DDOS-RSTFINFLOOD', 'DDOS-ICMP_FLOOD', 'DDOS-TCP_FLOOD',
       'DOS-UDP_FLOOD', 'DDOS-UDP_FLOOD', 'DDOS-SYN_FLOOD',
       'DNS_SPOOFING', 'MIRAI-GREIP_FLOOD', 'MIRAI-UDPPLAIN',
       'DDOS-PSHACK_FLOOD', 'DOS-SYN_FLOOD', 'DDOS-SYNONYMOUSIP_FLOOD',
       'DOS-TCP_FLOOD', 'BENIGN', 'DDOS-UDP_FRAGMENTATION',
       'DDOS-ACK_FRAGMENTATION', 'MIRAI-GREETH_FLOOD',
       'DDOS-ICMP_FRAGMENTATION', 'RECON-PORTSCAN', 'VULNERABILITYSCAN',
       'RECON-OSSCAN', 'RECON-HOSTDISCOVERY', 'MITM-ARPSPOOFING',
       'BACKDOOR_MALWARE', 'DOS-HTTP_FLOOD', 'DICTIONARYBRUTEFORCE',
       'DDOS-HTTP_FLOOD', 'DDOS-SLOWLORIS', 'BROWSERHIJACKING', 'XSS',
       'SQLINJECTION', 'COMMANDINJECTION', 'UPLOADING_ATTACK',
       'RECON-PINGSWEEP']
```

```python
import polars as pl
# Read for one attack from all the files and add then to a list.
# list=[ attack_file1, attack_file2,....]
i=1
balanced_df=[]


for name in attacks:
    print("Attack:",name)
                    # attack_file_count=[]
    for i in range(1,64):        
                        if(i<=9):
                            # using pandas for reading large file will waste time and storage so we will use polars to read only selected lines
                            # Some files may not have some attacks to handle that we need to put condition first
                            df=pl.scan_csv(f"/kaggle/input/labeled-ciciot-2023-dataset/Merged0{i}.csv").filter(pl.col('Label')==name)                                                                                      
                        else:
                            df=pl.scan_csv(f"/kaggle/input/labeled-ciciot-2023-dataset/Merged{i}.csv").filter(pl.col('Label')==name)
                            
                        df=df.collect() 
                        if df.is_empty():
                            continue
                        else:
                            current_count=len(df)
                            if current_count > 19:
                                df= df.sample(n=19,seed=0,shuffle=True)
                            else:
                                pass
                        df=df.to_pandas()
                        balanced_df.append(df)
      


# Then, from each file, we can get 19 rows randomly from each file. 
# And some files may not have some attacks, then we can add one attack of that attack from the reset of the files
# Then add all that mixed rows to a dataframe
```

```python
from sklearn.metrics import classification_report
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder


df=balanced_df
df.fillna({'Std':df['Std'].mean()},inplace=True)
df.fillna({'Variance':df['Variance'].mean()},inplace=True)
cols_to_drop = ['Number', 'IRC', 'IGMP', 'cwr_flag_number', 'ece_flag_number', 'Telnet', 'SMTP']
# Only drop if they actually exist in your dataframe
existing_cols_to_drop = [c for c in cols_to_drop if c in df.columns]
df.drop(columns=existing_cols_to_drop, inplace=True)
X=(df.select_dtypes(include='number'))
y=df['Label']

X.replace([np.inf, -np.inf], np.nan, inplace=True)

X.fillna({'Rate':0},inplace=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

DT=DecisionTreeClassifier(criterion='entropy',max_depth=25,min_samples_leaf=20,random_state=0)
DT.fit(X_train,y_train_enc)
y_predict=DT.predict(X_test)
print(classification_report(y_test_enc,y_predict,target_names=le.classes_))
```

The accuracy is not improving with this balanced dataset also. We can observe that decisiontree is predicting some attacks very well and some very poor due to high similarity of them with others. Possible solution : group the similar attacks so decisiontree can classifiy them easily.

```python
Attack_dictionary={"DDOS":[],"DOS":[],"MIRAI":[],"RECON":[],"BRUTEFORCE":[],"SPOOFING":[],"WEB_BASED":[],"BENIGN":[]}

def Labeling(df,unique):    
        for i in unique:
            word=[]
            index=df[df['Label']==i].index
            # print(index)
            words=i.split('-')
            # print("(*********************)")
            # print(i)
            # print(words[0])
            # print("(*********************)")
            
            if(words[0]=='DDOS'):
                Attack_dictionary['DDOS'].append(i)
                df.loc[index,'Label']='DDOS'
            elif(words[0]=='DOS'):
                Attack_dictionary['DOS'].append(i)        
                df.loc[index,'Label']='DOS'
                
            elif(words[0]=='MIRAI'):
                Attack_dictionary['MIRAI'].append(i)        
                df.loc[index,'Label']='MIRAI'
                
            elif(words[0]=='RECON' or words[0]=="VULNERABILITYSCAN"):
                Attack_dictionary['RECON'].append(i)      
                df.loc[index,'Label']='RECON'
                
            elif(i=="DICTIONARYBRUTEFORCE"):
                Attack_dictionary['BRUTEFORCE'].append(i) 
                
                df.loc[index,'Label']='BRUTEFORCE'
            elif(words[len(words)-1]== 'DNS_SPOOFING') or (words[len(words)-1]== 'ARPSPOOFING'):
                Attack_dictionary['SPOOFING'].append(i)      
                df.loc[index,'Label']='SPOOFING'
                
            elif(i=='BENIGN'):
                Attack_dictionary['BENIGN'].append(i)      
                df.loc[index,'Label']='Normal'
                
            else:
                Attack_dictionary['WEB_BASED'].append(i)      
                
                df.loc[index,'Label']='WEB_BASED'
Labeling(df,unique)
```

<h2>Data loading using Polars</h2>
<p>In this we are not loading the completing csv's but selecting the rows we need and adding them to dataframe this approach save time to loading large csv's</p>

```python

print("Adding data by Attach Class:")
balanced_df=[]
for Class,Subclasses in Attack_dictionary.items():
    print("Attack:",Class)
                    # attack_file_count=[]
    for i in range(1,64):
        for subclass in Subclasses:
                        #num_rows_from_each_file_by_each_subclass
                        num_rows=((11324/(len(Subclasses)))/63)
                        if(i<=9):
                            # using pandas for reading large file will waste time and storage so we will use polars to read only selected lines
                            # Some files may not have some attacks to handle that we need to put condition first
                            df=pl.scan_csv(f"/kaggle/input/labeled-ciciot-2023-dataset/Merged0{i}.csv").filter(pl.col('Label')==subclass)                                                                                      
                        else:
                            df=pl.scan_csv(f"/kaggle/input/labeled-ciciot-2023-dataset/Merged{i}.csv").filter(pl.col('Label')==subclass)                        
                        df=df.collect()
                        if df.is_empty():
                            continue
                        else:
                            current_count=len(df)
                            if current_count > int(num_rows):
                                df= df.sample(n=int(num_rows),seed=0,shuffle=True)
                            else:
                                pass
                        df=df.to_pandas()
                        balanced_df.append(df)
```
